### Step 0. Install and load
About 30 seconds in Colab.

```python
!pip install -q transformers

from transformers import AutoTokenizer
import pandas as pd
import matplotlib.pyplot as plt

# One multilingual tokenizer scores every language on the same scale.
tok = AutoTokenizer.from_pretrained("bert-base-multilingual-cased")
print("tokenizer loaded")
```

### Step 1. The same sentence in four languages
Meaning held constant: *Natural language processing is changing how businesses work.* If the meaning is the same, any difference in token count is a pure tax on the language, not on the message.

```python
sentences = {
    "English": "Natural language processing is changing how businesses work.",
    "Spanish": "El procesamiento del lenguaje natural está cambiando cómo trabajan las empresas.",
    "Hindi":   "प्राकृतिक भाषा संसाधन बदल रहा है कि व्यवसाय कैसे काम करते हैं।",
    "Chinese": "自然语言处理正在改变企业的工作方式。",
}
```

### Step 2. Count tokens and compute the tax
We use English as the baseline. A ratio of 2.0 means that language costs twice as many tokens for the same meaning.

```python
rows = []
for lang, sent in sentences.items():
    pieces = tok.tokenize(sent)
    rows.append({"language": lang, "characters": len(sent),
                 "tokens": len(pieces)})

df = pd.DataFrame(rows)
baseline = df.loc[df["language"] == "English", "tokens"].iloc[0]
df["cost_vs_english"] = (df["tokens"] / baseline).round(2)
df
```

### Step 3. See it
Colors are our course palette.

```python
bar_colors = ["#151A2E", "#B5A06A", "#B31A38", "#3C4A78"]
plt.figure(figsize=(7, 4))
plt.bar(df["language"], df["tokens"],
        color=bar_colors[: len(df)])
plt.ylabel("tokens for the same sentence")
plt.title("The token tax: same meaning, different cost")
for i, v in enumerate(df["tokens"]):
    plt.text(i, v + 0.3, str(v), ha="center")
plt.tight_layout()
plt.show()
```
