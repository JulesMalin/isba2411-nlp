### Step 0. Install and load the model
About 30 seconds in Colab.

```python
!pip install -q spacy
!python -m spacy download en_core_web_sm -q

import spacy
from spacy import displacy
from collections import Counter

nlp = spacy.load("en_core_web_sm")
print("model loaded")
```

### Step 1. The text
A default news-style paragraph is provided. In Step 4 you will swap in your own. Pick something messy for the rerun: a sports recap, a finance headline, or a tweet.

```python
text = (
    "Apple chief executive Tim Cook met European regulators in Brussels on Tuesday "
    "to discuss a proposed 2 billion euro fine. Shares of Apple fell 3 percent "
    "after the news, while the Nasdaq closed slightly higher."
)
doc = nlp(text)
```

### Step 2. Part-of-speech tags
`pos_` is the coarse universal tag (NOUN, VERB). `tag_` is the fine-grained Penn Treebank tag (NNP, VBD).

```python
pos_rows = [(t.text, t.pos_, t.tag_) for t in doc if not t.is_space]
for text_, pos, tag in pos_rows[:20]:
    print(f"{text_:<14} {pos:<8} {tag}")
```

### Step 3. Named entities
Each entity has text, a label, and a position. The displacy view colors them inline using our course palette.

```python
for ent in doc.ents:
    print(f"{ent.text:<22} {ent.label_:<8} ({spacy.explain(ent.label_)})")
```

```python
colors = {
    "PERSON": "#B5A06A", "ORG": "#B31A38",
    "GPE": "#3C4A78", "MONEY": "#B31A38", "DATE": "#B5A06A",
}
options = {"colors": colors}
displacy.render(doc, style="ent", jupyter=True, options=options)
```

### Step 4. POS distribution
How noun-heavy is this text? News tends to be dense with proper nouns.

```python
counts = Counter(t.pos_ for t in doc if not t.is_space and not t.is_punct)
for pos, n in counts.most_common():
    bar = "#" * n
    print(f"{pos:<8} {n:>2}  {bar}")
```
